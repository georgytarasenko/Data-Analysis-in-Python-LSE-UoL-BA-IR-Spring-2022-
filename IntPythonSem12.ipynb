### Correlation

Correlation is a statistical term describing the degree to which two variables move in coordination with one another. If the two variables move in the same direction, then those variables are said to have a positive correlation. If they move in opposite directions, then they have a negative correlation.

![Venus_flytrap_taxonomy.jpg](https://cdn.scribbr.com/wp-content/uploads/2021/08/01-correlation-types-1024x415.png)



In statistics, the **Pearson correlation coefficient** is a measure of linear correlation between two sets of data. It is the ratio between the covariance of two variables and the product of their standard deviations; thus it is essentially a **normalized** measurement of the covariance, such that the result always has a value between `−1` and `1`.

```python
cor = midwest[['area', 'poptotal', 'popdensity', 'popwhite', 'popblack']].corr()

cor.round(2)
```

```python
cmap = sns.diverging_palette(0, 250, as_cmap=True)
sns.heatmap(cor, cmap=cmap, square=True, linewidth=0.2)
```

### Apply method
You remember that when we are iterating `list`, we take each value as `i` (single value) (by the way, what we take as a single value when iterating `dict`?). Let's check what we take as a single value as we iterate `pd.DataFrame`:

But what if we want to apply some function to each row? In this case `.apply()` method (of `pd.DataFrame`) comes:

```python
midwest1st5.apply(lambda row: print(row), axis=1) # anonymous function lambda: select the name of 'each varible' and then
# use it
```

`axis=1` argument allows us to iterate through rows while if we set `axis=0` (default) we iterate columns (not column name but columns):

```python
midwest1st5.apply(lambda column: print(column), axis=0)
```

What is one row of `pd.DataFrame`? Let's see:

```python
midwest1st5.iloc[0, :] # it is pd.Series where column names have become indices
```

Suppose now that we want to create a new variable - `popdensity_cat` - which takes values `low density` (< lower quartile), `medium density` (between lower quartile and median), `high density` (between median and higher quartile) and `very high density` (> higher quartile). In this case we should take each row, select `popdensity`, transform it and return a new value. Here `.apply()` in combination with custom function comes:

```python
median = midwest['popdensity'].median()
lQ = midwest['popdensity'].quantile(0.25)
hQ = midwest['popdensity'].quantile(0.75)
```

```python
def categorize_density(row, median, lQ, hQ): # as we iterate rows using apply
    density = row['popdensity']
    
    if density >= hQ:
        return 'very high density'
    elif density < hQ and density >= median:
        return 'high density'
    elif density < median and density >= lQ:
        return 'medium density'
    else:
        return 'low density'
```

```python
midwest['popdensity_cat'] = midwest.apply(lambda row: categorize_density(row, median, lQ, hQ), axis=1)
midwest.head()
```
